```python
# create 4 random vectors
r1 = np.array([0, 3, -3, 0])
r2 = np.array([2, 6,- 2, 2])
r3 = np.array([-2, 1, -3, 0])
r4 = np.array([2, -4, -2, -4])
```

```python
print("the first rsndom vector is ", r1)
```

# lets process the first vector

```python
# choose r1
# make a temporary vector t1 = r1
t1 = r1
print(t1)
```

```python
# normalize t1, and save it as our nice vector u1
n1 = np.sqrt(t1.dot(t1))
print(n1)
# lets form u1
u1 = 1/n1 * t1
print("our first nice vector is u1: ", u1)
```

```python
# sanity check:
# size ov u1 is 1, and it is orthogonal to other nice vectors
print(np.sqrt(u1.dot(u1)))
```

# process the secont vector r2

```python
# choose r2
# make a temporary vector t2 = r2 - (r2 . u1) u1
t2 = r2 - r2.dot(u1) * u1
print(t2)
```

```python
# normalize t2 and save it as u2
n2 = np.sqrt(t2.dot(t2))
print(n2)
# lets form u2
u2 = 1/n2 * t2
print("our first nice vector is u2: ", u2)
```

```python
# sanity check
# size of u2 is 1, and it is orthogonal to other nice vectors
print(np.sqrt(u2.dot(u2)))
print(u1.dot(u2))
```

# process third vector r3

```python
# choose r3
# make a temporary vector t3 = r3 - (r3 . u1) u1 - (r3 . u2)u2
t3 = r3 - r3.dot(u1) * u1 - r3.dot(u2) * u2
print(t3)
```

```python
# normalize t3 and save it as u3
n3 = np.sqrt(t3.dot(t3))
print(n3)
# lets form u3
u3 = 1/n3 * t3
print("our first nice vector is u3: ", u3)
```

```python
# sanity check
# size of u3 is 1, and it is orthogonal to other nice vectors
print(np.sqrt(u3.dot(u3)))
print(u1.dot(u3))
```

# process 4th vector r4

```python
# choose r4
# make a temporary vector t4 = r4 - (r4 . u1) u1 - (r4 . u2)u2 - (r4 . u3)u3
t4 = r4 - r4.dot(u1) * u1 - r4.dot(u2) * u2 - r4.dot(u3) * u3
print(t4)
```

```python
# normalize t4 and save it as u4
n4 = np.sqrt(t4.dot(t4))
print(n4)
# lets form u3
u4 = 1/n4 * t4
print("our first nice vector is u4: ", u4)
```

```python
# sanity check
# size of u4 is 1, and it is orthogonal to other nice vectors
print(np.sqrt(u4.dot(u4)))
print(u1.dot(u4))
```
